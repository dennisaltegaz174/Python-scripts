```python
# Regression
# simmple linear model where dependent  variable is tenure and independent variable  is 'monthlycharges'
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
# Build  a simple linear regresion model
data =  pd.read_csv("C:\\Users\\adm\\Documents\\Datasets\\Customer_churn\\WA_Fn-UseC_-Telco-Customer-Churn.csv")
x = data.loc[:,['tenure']].values
y =  data.loc[:,['MonthlyCharges']].values
```

```python
# Divide the data into train and test data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=0)
```

```python
# model  building and making prediction 
from sklearn.linear_model import LinearRegression
SimpleLinearRegression = LinearRegression()
SimpleLinearRegression.fit(x_train,y_train)
```

```python
# After predicting the values find the mean square error
y_pred = SimpleLinearRegression.predict(x_test)
y_pred
```

```python
y_test
```

```python
# Root mean square error
from sklearn.metrics import mean_squared_error
mse =  mean_squared_error(y_pred,y_test)
rmse= np.sqrt(mse)
rmse
#  a high value of rmse (more than 10 or 11 ) means our model is not goood and cannot be used for prediction
```

```python
# Logistic Regression
# Build a simple logistic model where dependennt variable is 'churn' and indepennt variable is 'MonthlyCharges'
# split data into in 65:25rratio
```

```python
# model building
y = data.loc[:,['Churn']].values
x = data.loc[:,['MonthlyCharges']].values
```

```python
# Divide data into test and training sets
x_train, x_test, y_train, y_test = train_test_split(x,y,train_size=0.65,random_state = 0)
# Build  model on train set and predict values on test set
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
```

```python
# confusion matrix
# True positive(TP): Observation is positive and is predicted to be positive.
# False Negative(FN):  Observation is positive  but its predicted negative.
# True Negative(TN): Observation is negative  and is predicted to be negative
# False positive(FP): Observation is negative but is predicted to be positive
#         Predicted
#actual TP    FN
#       FP    TN
```

```python
# Build a multiple regression model with churn as the dependent variable and independent variables are tenure and 'monthly charges'
x =  data.loc[:,['MonthlyCharges', 'tenure']].values
y = data.loc[:,['Churn']].values
```

```python
# Divide data into training and test
x_train, x_test,y_train, y_test = train_test_split(x,y,train_size = 0.80,random_state = 0)
```

```python
# Model buildingb and prediction
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
```

```python
# Confusion matrix
print(confusion_matrix(y_test,y_pred))
```
